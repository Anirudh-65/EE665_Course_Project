```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
DATA_PROCESSED = Path("data_processed")

df = pd.read_csv(DATA_PROCESSED/"battery_master_dataset.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
df = df.dropna()

df = df[df["voltage"] > 0]
df = df[df["current"].abs() < 100]

print("After cleaning:", df.shape)
```

```python
print(df["dataset"].value_counts())
```

```python
print(df["dataset"].value_counts())
```

```python
df.groupby("dataset")["cycle"].nunique()
```

```python
df_aug = df.copy()

df_aug["voltage"] = df_aug["voltage"] + \
    np.random.normal(0, 0.003, len(df_aug))
```

```python
features = df.groupby(["cell_id","cycle","dataset"]).agg({
    "voltage":["mean","min","max","std"],
    "current":["mean"],
    "temperature":["mean"],
})

features.columns = ["_".join(col) for col in features.columns]
features = features.reset_index()

features.head()
```

```python
features.to_csv(DATA_PROCESSED/"battery_features_week2.csv", index=False)

print("Saved battery_features_week2.csv")
```
